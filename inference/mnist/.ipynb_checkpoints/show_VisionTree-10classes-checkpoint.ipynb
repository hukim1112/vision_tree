```python
import pandas as pd
import plotly
from utils.visualization import Feature, Vision_tree

L3,L2,L1,L0 = np.load("10classes/3.npy"),\
              np.load("10classes/2.npy"),\
              np.load("10classes/1.npy"),\
              np.load("10classes/0.npy")

vt = Vision_tree('mnist', [L1,L2,L3], 4)
vt.create_layers()
nodes, connections = vt.df_tree()
```

```python
import plotly.graph_objects as go


fig = go.Figure(data=[go.Scatter3d(x=nodes['x'], y=nodes['y'], z=nodes['z'],
                                   mode='markers', marker=dict(
                                                            size=6,
                                                            color=nodes['color'],   # set color to an array/list of desired values
                                                            colorscale='Viridis',   # choose a colorscale
                                                            opacity=1.0))])

for connection in connections.values:
    x1,y1,z1,x2,y2,z2,color = connection
    fig.add_trace(go.Scatter3d(
            x=[x1,x2],
            y=[y1,y2],
            z=[z1,z2],
            mode='lines',
            line=dict(
        color=color,
        width=2
    )))

fig.show()
```

```python
kld = tf.keras.losses.KLDivergence()
diversity = {}
for L in [3,2,1]:
    diversity[L] = []
    print(L)
    for f in vt.tree[L]:
        Q = f.inferiors
        P = tf.ones_like(Q)/len(Q)
        similarity = kld(P,Q).numpy()
        diversity[L].append(similarity)
```

```python
x = []
for L in diversity:
    #print(diversity[L])
    print(np.mean(diversity[L]))
    x.append(np.mean(diversity[L]))
```

```python
plt.plot([1,2,3],x[::-1], 'ro-')
plt.xlabel('lower layer index')
plt.ylabel('Kull-back leibler')
```
